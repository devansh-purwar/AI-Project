```python
df = pd.read_csv('../input/twitter-data/newData.csv')
positive =  ['😊','😉','😀','😃','🌚','😇','🎉','😅','😆','😎','😏','😈','😋','😜','😝','😌','😂','😁','😄','💕','💖','💗','💙','💛','💜','💚','💞','😍','😘','😢','✅','💯','🔥','👍','👏','👌','👀','💪','🙏','🙌','🙈','✨','🎶','👇','👑','😬','😱','😳']
negative  =  ['😭','😡','😤','😑','😒','💀','😣','😩','😪','😫','💔','😴','😐','😔','😕','😞']
def sentiment_map(emoji):
    if emoji in positive:
        return 1
    else :
        return 0
df['sentiment'] = df['label_to_emoji'].map(sentiment_map)
```

```python
df  = df.drop(columns = ['Unnamed: 0'])
```

```python
import emoji
def to_description(emo):
    return emoji.demojize(emo)
```

```python
df['emoji_description'] = df['label_to_emoji'].map(to_description)
```

```python
def clean_desc(des):
    return des.replace('_'," ").replace(':',"")
df['emoji_description'] = df['emoji_description'].map(clean_desc)
```

```python
df["tweet"] = df[["pre_punc", "emoji_description"]].apply(" ".join, axis=1)
```

Predecting sentiment, 7 diff emoji category classes and each class with emoji description in tweet

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df.tweet, 
    df.sentiment, 
    test_size=0.2, # 20% samples will go to test dataset
    random_state=2022,
    stratify=df.sentiment
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
