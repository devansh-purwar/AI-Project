```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
Uinput  = ['You love hurting me, huh?','This is the shit!','My flight is delayed.. amazing.','What is happening to me??',"I know good movies, this ain't one"]
X = np.asarray(Uinput) 
emoji_raw = open('../input/model-demostration/us_mapping.txt','r',encoding="utf8")
emojis=[]
for sentence in emoji_raw:
    sentence = sentence.rstrip()
    emojis.append(sentence) 
emoji_dict={}
for e in emojis:
    idx = int(e.split()[0])
    emoji = e.split()[1]
    emoji_dict[idx] = emoji
from keras.preprocessing.sequence import pad_sequences
tokenizer = pickle.load(open("../input/model-demostration/tweet_tokenizer",'rb'))
def preprocess_text(X):
    max_len=40
    X_seqs = tokenizer.texts_to_sequences(X)
    X_seqs_pd = pad_sequences(X_seqs, truncating="pre", padding="pre", maxlen=max_len)
    return X_seqs_pd
from tensorflow import keras
model = keras.models.load_model("../input/model-demostration//BLSTM.h5", compile=False)
model.summary()
X_sequences = preprocess_text(X)
predictions = np.argmax(model.predict(X_sequences), axis=1)
predictions = [emoji_dict[idx] for idx in predictions]
results = pd.DataFrame({"Tweet":X})
results["Predicted"] = predictions
results
```
