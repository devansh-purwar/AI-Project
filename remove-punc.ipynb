```python
df_train =  pd.read_csv('../input/twitter-separated/train_twitter.csv')
df_test =  pd.read_csv('../input/twitter-separated/test_twitter.csv')
```

```python
import string

def punc(s):
    table = str.maketrans(dict.fromkeys(string.punctuation))  # OR {key: None for key in string.punctuation}
    new_s = s.translate(table)   
    return new_s
```

```python
df_train['description'] = df_train['description'].map(punc)
df_test['0'] = df_test['0'].map(punc)
```
